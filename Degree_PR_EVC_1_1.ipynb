```python
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


df_read_tp  = pd.read_csv('https://raw.githubusercontent.com/rahmanziaur/TestGraphTheory/main/DegreeTest.csv')

read_label = df_read_tp['Label']
read_tp    = df_read_tp['Degree']

y_pos = range(len(read_label))

# plt.bar(y_pos, heights)
# # Rotation of the bars names
# plt.xticks(y_pos, bars, rotation=90)

sns.barplot(x=read_label, y=read_tp)
# Add labels to your graph
plt.xlabel('Packages')
plt.ylabel('Degree')
plt.xticks(y_pos, read_label, rotation=90)
plt.title("PyPI packages with Degree Centrality higher than 200")
plt.savefig('my_plot.png')
plt.show()
```
